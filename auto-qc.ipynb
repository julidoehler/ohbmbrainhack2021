```python
import csv

row_list = [['/home/doehlerj/Dokumente/brainhack2021/data_rawBIDS/BMB/sub-0003001/ses-1/anat/sub-0003001_ses-1_run-1_T1w.nii.gz','/home/doehlerj/Dokumente/brainhack2021/data_mriqc/sub-0003001/ses-1/anat/sub-0003001_ses-1_run-1_T1w.csv'],
           ['/home/doehlerj/Dokumente/brainhack2021/data_rawBIDS/BMB/sub-0003002/ses-1/anat/sub-0003002_ses-1_run-1_T1w.nii.gz','/home/doehlerj/Dokumente/brainhack2021/data_mriqc/sub-0003002/ses-1/anat/sub-0003002_ses-1_run-1_T1w.csv']]

filepaths_name = 'filepaths'
```

```python
def paths_to_csv(filepaths_name,row_list):
    with open(filepaths_name + '.csv','w',newline='') as file:
        writer = csv.writer(file,delimiter=',')
        writer.writerows(row_list)
```

```python
path_data = 'data_mriqc'
path_image = 'data_rawBIDS/BMB'

row_list_anat = []
row_list_func = []

for subj in subject_list:
    
    files = glob.glob(os.path.join(path_data,subj,session))
    
    if len(files) == 0:
        print('skip ', subj)
        continue
    
    print('Processing ', subj)
    session='ses-1'
    for mri in ['anat','func']:
        if mri == 'anat':
            csv_file_name = '{}_{}_run-1_T1w'.format(subj,session)
            csv_file = os.path.join(path_data,subj,session,mri,csv_file_name)
            image_file = os.path.join(path_image,subj,session,mri,csv_file_name)
            row_list_anat.append([image_file + '.nii.gz',csv_file + '.csv'])
        else:
            files = glob.glob(os.path.join(path_data,subj,session,mri)+'/*.json')
            for j in range(1,len(files)+1):
                csv_file_name = '{}_{}_task-rest_run-{}_bold'.format(subj,session,j)
                csv_file = os.path.join(path_data,subj,session,mri,csv_file_name)
                image_file = os.path.join(path_image,subj,session,mri,csv_file_name)
                row_list_func.append([image_file + '.nii.gz',csv_file + '.csv'])
                
```
